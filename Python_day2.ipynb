# 반복문 while, for

- 파이썬에서의 반복문의 while or for 두 가지로 사용할 수 있음.
while 로 구현할 수 있는 모든 반복문은 for 로 도 구현할 수 있고 그 반대의 경우도 100% 가능함.
- 그러나, while 은 반복 횟수가 정확하지 않을 경우 주로 사용되고 for 문은 반복횟수가 명확할 때 주로 사용됨. 
  - 반복을 100번 이상 돌리게 되면 while을 쓰고, 명확하게 20번 돌린다는 내용이라면 for 문을 사용하는 것임.
